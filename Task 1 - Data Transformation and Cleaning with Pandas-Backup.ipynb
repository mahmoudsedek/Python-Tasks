### Step 0
Load the energy data from the file `Energy Indicators.xls`, which is a list of indicators of [energy supply and renewable electricity production](Energy%20Indicators.xls) from the [United Nations](http://unstats.un.org/unsd/environment/excel_file_tables/2013/Energy%20Indicators.xls) for the year 2013, and should be put into a DataFrame with the variable name of **energy**.

Keep in mind that this is an Excel file, and not a comma separated values file. Also, make sure to exclude the footer and header information from the datafile. The first two columns are unneccessary, so you should get rid of them, and you should change the column labels so that the columns are:

`['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']`

Convert `Energy Supply` to gigajoules (there are 1,000,000 gigajoules in a petajoule). For all countries which have missing data (e.g. data with "...") make sure this is reflected as `np.NaN` values.

Rename the following list of countries (for use in later questions):

```"Republic of Korea": "South Korea",
"United States of America": "United States",
"United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
"China, Hong Kong Special Administrative Region": "Hong Kong"```

There are also several countries with numbers and/or parenthesis in their name. Be sure to remove these, 

e.g. 

`'Bolivia (Plurinational State of)'` should be `'Bolivia'`, 

`'Switzerland17'` should be `'Switzerland'`.

````python
#importing Packages
import pandas as pd
import numpy as np
import re
#Loading the Excel File with the appropiate aspects
file_path="/home/mahmoud/ITI_Python_for_Data_Management/Task 1/data/Energy Indicators.xls"

energy=pd.read_excel(file_path, skiprows=17 ,nrows=227, usecols="C:F", na_values='...') #specify used cols and no. of rows, skipping header, and nan values
#renaming coloumns
energy.rename(columns={"Unnamed: 2":'Country',"Petajoules":'Energy Supply',"%":'% Renewable',"Gigajoules":"Energy Supply per Capital"},inplace = True)

energy.head(7)
````

````python
energy.info()
````

````python
#converting 'Energy Supply' to Gigajoules (1,000,000 gigajoules in a petajoule)
# energy.replace('...', np.nan,inplace = True)
energy['Energy Supply'] = energy['Energy Supply'].apply(lambda x: x*1000000)
````

````python
energy.head(7)
````

````python
energy.info()
````

````python
#function to clean Countries from paranthes, numbers and renaming some Countries
def clean_Parenthes(country):
    index = country.find('(')
    if(index != -1):
        return country[:index].strip()
    return country.strip()

def clean_Number(country):
    country = re.sub("\d+"," ",country)
    return country.strip()

def rename_country(country):
    if(re.search("^Republic of Korea",country)!=None):
        return "South Korea"
    elif(re.search("^United States of America",country)!=None):
        return "United States"
    elif(re.search("^United Kingdom of Great Britain and Northern Ireland",country)!=None):
        return "United Kingdom"
    elif(re.search("^China, Hong Kong Special Administrative Region",country)!=None):
        return "Hong Kong"
    elif(re.search("^The former Yugoslav Republic of Macedonia",country)!=None):
        return "North Macedonia"
    elif(re.search("^Democratic People's Republic of Korea",country)!=None):
        return "North Korea"
    return country
    
#corresponding manual code for the prev function

# energy['Country'] = energy['Country'].str.replace(r"\(.*\)","")
# energy['Country'] = energy['Country'].str.replace("[0-9()]+$", "")
# energy["Country"].replace(["Republic of Korea","United States of America",
#                                "United Kingdom of Great Britain and Northern Ireland",
#                                "China, Hong Kong Special Administrative Region",
#                           'Bolivia (Plurinational State of)',
#                            'Switzerland17','The former Yugoslav Republic of Macedonia',
#                           "Democratic People's Republic of Korea"],
                          
#                               ["South Korea","United States","United Kingdom","Hong Kong",
#                                'Bolivia','Switzerland','North Macedonia','North Korea']
#                           ,inplace = True)

energy['Country']=energy['Country'].apply(clean_Parenthes).apply(clean_Number).apply(rename_country)
energy.sample(7)  #show diff. samples and make sure the all countries have changed correctly
````

````python
# energy.isnull().sum()
# energy.info()
#output the file
energy.to_excel('/home/mahmoud/ITI_Python_for_Data_Management/Task 1/data/output/Energy Indicators2.xls')
energy.to_csv('/home/mahmoud/ITI_Python_for_Data_Management/Task 1/data/output/Energy Indicators2.csv')
````
